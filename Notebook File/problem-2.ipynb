### Loading Datasets from CSV File

```python
purchase_df = pd.read_csv('/kaggle/input/robi-datathon-30-qualifier-problem-2/purchase.csv')
purchase_df.head()
```

```python
boxes_df = pd.read_csv('/kaggle/input/robi-datathon-30-qualifier-problem-2/boxes.csv')
boxes_df.head()
```

```python
### Exploration and Feature Engineering
```

```python
# Displaying information about the purchase_df DataFrame
purchase_df.info()
```

```python
# Displaying information about the boxes_df DataFrame
boxes_df.info()
```

```python
# remove all the rows that contain a missing value
purchase_df.dropna(inplace=True)
```

```python
# Convert 'BOX_ID' column in 'purchase' dataset to int64
purchase_df['BOX_ID'] = purchase_df['BOX_ID'].astype(int)
```

```python
#merge the datasets based on 'BOX_ID'
merged_df = pd.merge(purchase_df, boxes_df, on='BOX_ID')
merged_df
```

```python
merged_df.info()
```

```python
# Calculating the number of missing values in each column of the merged DataFrame
merged_df.isnull().sum()
```

```python
import datetime  # Importing datetime module for date manipulation

# Converting 'PURCHASE_DATE' column to datetime format using the specified format
merged_df['PURCHASE_DATE'] = pd.to_datetime(merged_df['PURCHASE_DATE'], format='%d/%m/%Y')
```

```python
# Get the oldest and latest dates
oldest_date = merged_df['PURCHASE_DATE'].min()
latest_date = merged_df['PURCHASE_DATE'].max()

print("Oldest Date:", oldest_date)
print("Latest Date:", latest_date)
print(latest_date-oldest_date)
```

```python
# Extracting the day of the month from the 'PURCHASE_DATE' column
merged_df['DAY_OF_MONTH'] = merged_df['PURCHASE_DATE'].dt.day

# Calculating the cost by multiplying 'BOX_COUNT' and 'UNIT_PRICE'
merged_df['COST'] = merged_df['BOX_COUNT'] * merged_df['UNIT_PRICE']

# Displaying the first few rows of the merged DataFrame with the new columns added
merged_df.head()
```

### Encoding

```python
# Getting the unique values of the 'QUALITY' column in the merged DataFrame
merged_df['QUALITY'].unique()
```

```python
# Defining the order of the QUALITY levels
QUALITY_ORDER = ['Standard', 'Premium']
```

```python
from sklearn.preprocessing import OrdinalEncoder  # Importing OrdinalEncoder from scikit-learn for ordinal encoding

# Creating an instance of OrdinalEncoder with specified categories
o_encoder = OrdinalEncoder(categories=[QUALITY_ORDER])

# Displaying the OrdinalEncoder instance
o_encoder
```

```python
# Fitting the OrdinalEncoder to the 'QUALITY' column of the merged DataFrame
o_encoder.fit(merged_df[['QUALITY']])
```

```python
# Transforming the 'QUALITY' column of the merged DataFrame using the fitted OrdinalEncoder
encoded_data = o_encoder.transform(merged_df[['QUALITY']])
```

```python
# Updating the 'QUALITY' column of the merged DataFrame with the encoded data
merged_df['QUALITY'] = pd.DataFrame(data=encoded_data, columns=['QUALITY'])
```

```python
merged_df['QUALITY'].dtype
```

```python
# Calculating the number of unique MAGIC_KEY values in the merged DataFrame
len(merged_df['MAGIC_KEY'].unique())
```

```python
# Perform one-hot encoding for DELIVERY_OPTION
delivery_option_encoded = pd.get_dummies(merged_df['DELIVERY_OPTION'])
```

```python
# Concatenate the one-hot encoded columns with the original dataframe
merged_df = pd.concat([merged_df, delivery_option_encoded], axis=1)
```

```python
# Drop the original DELIVERY_OPTION column
merged_df.drop('DELIVERY_OPTION', axis=1, inplace=True)
```

```python
merged_df.head()
```

```python
# Adding a new column 'Buy_in_first_15D' to the merged DataFrame
# indicating whether the purchase was made in the first 15 days of the month
merged_df['Buy_in_first_15D'] = merged_df['DAY_OF_MONTH'] <= 15
```

```python
merged_df.head()
```

```python
# Counting the number of occurrences of each BOX_ID value in the merged DataFrame
merged_df['BOX_ID'].value_counts()
```

```python
# Setting the 'MAGIC_KEY' column as the index of the merged DataFrame
merged_df.set_index('MAGIC_KEY', inplace=True)
```

```python
merged_df.head()
```

```python
# Get the list of magic keys from problem2_df
problem_keys = problem2_df['MAGIC_KEY'].tolist()

# Remove rows corresponding to these keys from the training data
filtered = merged_df[merged_df.index.isin(problem_keys)]
```

```python
print("Shape of filtered:", filtered.shape)
```

### Train-Test Split

```python
# Separate features and target variable
X_train = filtered.drop(['PURCHASE_DATE', 'BOX_ID', 'DAY_OF_MONTH','COST'], axis=1)
y_train = filtered[['BOX_ID']]
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
```

```python
# Get the list of columns
columns = list(X_train.columns)

# Remove QUALITY from its current position
columns.remove('QUALITY')

# Insert QUALITY at the fifth position
columns.insert(4, 'QUALITY')

# Reorder the columns in the DataFrame
X_train = X_train[columns]
```

```python
X_train.head()
```

```python
# Converting boolean columns to integers (0 and 1)
X_train['Delivery from Collection Point'] = X_train['Delivery from Collection Point'].astype(int)
X_train['Home Delivery - CoD'] = X_train['Home Delivery - CoD'].astype(int)
X_train['Home Delivery - Digital Payment'] = X_train['Home Delivery - Digital Payment'].astype(int)
X_train['Buy_in_first_15D'] = X_train['Buy_in_first_15D'].astype(int)

# Displaying the modified DataFrame
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Specify the columns to be standardized
columns_to_scale = ['BOX_COUNT', 'MILK', 'MEAT', 'UNIT_PRICE']

# Scale the specified columns
X_train[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])

# Displaying the modified DataFrame
X_train.head()
```

```python
X_train.head()
```

```python
y_train.head()
```

**The test data will be of problem2_df but all their BOX_COUNT will be set as 1 (0.0 after standardized) and Buy_in_first_15D will be all set to 1 (True). As in the question it is explicitly mentioned that they will buy only 1 box and in the first 15 days of march. For Other features mean value and mode value will be chosen.**

```python
# Filtering the DataFrame based on the condition where 'BOX_COUNT' is equal to 0.0
test_filter = X_train[(X_train['BOX_COUNT'] == 0.0)]

# Displaying the filtered DataFrame
test_filter
```

```python
# Counting the number of unique indices in the filtered DataFrame
test_filter.index.nunique()
```

```python
# Getting the unique values of 'BOX_COUNT' in the filtered DataFrame
test_filter['BOX_COUNT'].unique()
```

```python
# Calculating the mean of the first four columns for each unique index in the filtered DataFrame
group1 = test_filter.iloc[:, 0:4].groupby(level=0).mean()
```

```python
group1
```

```python
# Calculating the mode of the remaining columns for each unique index in the filtered DataFrame
group2 = test_filter.iloc[:, 4:].groupby(level=0).agg(lambda x: x.mode().iloc[0])
```

```python
group2
```

```python
# Concatenating group1 and group2 along the columns (axis=1) to form the X_test DataFrame
X_test = pd.concat([group1, group2], axis=1)
```

```python
X_test['Buy_in_first_15D'] = 1
```

```python
X_test
```

## RandomForestClassifie

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust the number of trees (n_estimators)

# Train the model on the training data
rf_classifier.fit(X_train, y_train)
```

```python
# Use the trained model to make predictions on the test data
rf_pred = rf_classifier.predict(X_test)
```

```python
rf_pred
```

```python
# Creating a DataFrame with 'MAGIC_KEY' from X_test index and 'BOX_ID' predictions
predictions_df = pd.DataFrame({'MAGIC_KEY': X_test.index, 'BOX_ID': rf_pred})
```

```python
predictions_df
```

```python
# Merging problem2_df with predictions_df on 'MAGIC_KEY'
submission = pd.merge(problem2_df, predictions_df, on='MAGIC_KEY')

# Displaying the merged DataFrame
submission
```

```python
# Saving the submission DataFrame to a CSV file without including the index
submission.to_csv('Submission.csv', index=False)
```
